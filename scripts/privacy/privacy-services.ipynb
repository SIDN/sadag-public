```python
legacyTotal = legacyTotal.groupby(legacyTotal.index).sum()
legacyPrivacy = legacyPrivacy.groupby(legacyPrivacy.index).sum()
newTotal = newTotal.groupby(newTotal.index).sum()
newPrivacy = newPrivacy.groupby(newPrivacy.index).sum()
dfJoined = pd.concat([newTotal,newPrivacy,legacyTotal, legacyPrivacy], axis=1)
dfJoined["newgtld_percent"] = (dfJoined["privacy_count_newgtld"] / dfJoined["total_count_newgtld"]) * 100.0
dfJoined["legacygtld_percent"] = (dfJoined["privacy_count_legacygtld"] / dfJoined["total_count_legacygtld"]) * 100.0

dfJoined["legacy_ratio"] = (dfJoined["privacy_count_legacygtld"] / dfJoined["total_count_legacygtld"]) * 100.0

dfJoined.fillna(value=0, inplace=True)
dfJoined.round(2).to_csv('privacy_services_overall.csv')
dfJoined
```
