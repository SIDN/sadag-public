```python
def processData(fileInNew, fileInLegacy, fileOut):
    dfLegacy = pd.read_csv(fileInLegacy, names=["legacy_country","legacy_count"]) # index_col="country"
    dfNew = pd.read_csv(fileInNew,names=["newgtld_country","newgtld_count"])
    #calc ratios
    dfLegacy["legacy_rate"] = (dfLegacy["legacy_count"] / dfLegacy.sum(axis=0)["legacy_count"]) * 100.0
    dfLegacy["legacy_rate/10k"] = (10000 *dfLegacy["legacy_count"]) / dfLegacyCountrySize.loc[dfLegacy["legacy_country"]].values[0]
    dfLegacy = dfLegacy.sort_values("legacy_rate/10k", ascending=False)
    dfLegacy['legacy_country'] = dfLegacy['legacy_country'].apply(lambda x: x.title())
    dfLegacy["legacy_count"] = dfLegacy["legacy_count"].astype(int)
    dfLegacy['legacy_count'] = dfLegacy['legacy_count'].apply(lambda x: '{0:,d}'.format(x))

    dfNew["newgtld_rate"] = (dfNew["newgtld_count"] / dfNew.sum(axis=0)["newgtld_count"]) * 100.0
    dfNew["newgtld_rate/10k"] = (10000 *dfNew["newgtld_count"]) / dfNewCountrySize.loc[dfNew["newgtld_country"]].values[0]
    dfNew = dfNew.sort_values("newgtld_rate/10k", ascending=False) 
    dfNew['newgtld_country'] = dfNew['newgtld_country'].apply(lambda x: x.title())
    dfNew["newgtld_count"] = dfNew["newgtld_count"].astype(int)
    dfNew['newgtld_count'] = dfNew['newgtld_count'].apply(lambda x: '{0:,d}'.format(x))

    dfResult = pd.concat([dfNew,dfLegacy], axis=1)
    #dfResult = dfNew
    #let index start at 1
    dfResult.index += 1 
    dfResult = dfResult[0:10]
    dfResult.fillna(value=0, inplace=True)
    #dfResult["newgtld_count"] = dfResult["newgtld_count"].astype(int)

    
    dfResult.round(2).to_csv(fileOut, sep='&')

    return dfResult.round(2)
```

```python
dfTopCountries = pd.read_csv('top-registrar-countries.csv', header=0,index_col=0)
dfTopCountries['share'] = dfTopCountries['registrars'] /  dfTopCountries['registrars'].sum() * 100.0
dfTopCountries = dfTopCountries.round(2)
dfTopCountries[0:10].to_csv('top-registrar-countries-share.csv')
dfTopCountries[0:10]
```

```python
#domains per country
dfTopCountriesNew = pd.read_csv('select-country-size-newgtld.csv', header=0,index_col=0)
dfTopCountriesNew['share'] = dfTopCountriesNew['domains'] /  dfTopCountriesNew['domains'].sum() * 100.0
dfTopCountriesNew = dfTopCountriesNew.round(2)
dfTopCountriesNew.sort_values("domains")
dfTopCountriesNew.reset_index(inplace=True)
dfTopCountriesNew['country'] = dfTopCountriesNew['country'].apply(lambda x: x.title())
dfTopCountriesNew['domains'] = dfTopCountriesNew['domains'].apply(lambda x: '{0:,d}'.format(x))
dfTopCountriesNew[0:10].to_csv('country-domain-count-new-gtld-share.csv')
dfTopCountriesNew[0:10]
```

```python
#legacy domains per country
dfTopCountriesLegacy = pd.read_csv('select-country-size-legacygtld.csv', header=0,index_col=0)
dfTopCountriesLegacy['share'] = dfTopCountriesLegacy['domains'] /  dfTopCountriesLegacy['domains'].sum() * 100.0
dfTopCountriesLegacy = dfTopCountriesLegacy.round(2)
dfTopCountriesLegacy.sort_values("domains")
dfTopCountriesLegacy.reset_index(inplace=True)
dfTopCountriesLegacy['country'] = dfTopCountriesLegacy['country'].apply(lambda x: x.title())
dfTopCountriesLegacy['domains'] = dfTopCountriesLegacy['domains'].apply(lambda x: '{0:,d}'.format(x))
dfTopCountriesLegacy[0:10].to_csv('country-domain-count-legacy-gtld-share.csv')
dfTopCountriesLegacy[0:10]
```

```python
dfResult = pd.concat([dfTopCountriesNew[0:10],dfTopCountriesLegacy[0:10]], axis=1)
dfResult.index += 1 
dfResult.to_csv('country-domain-count-combined-share.csv', sep='&')
dfResult
```
