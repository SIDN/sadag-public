```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import re
```

```python
#1 read csv file
#2 normalize registar names by removing non alpha chars
#3 create index on normalized name
#4 group by normalized names
#5 join all dataframes into single dataframe (join on normalized registrar name)

#1
dfRegistrarAllDomains = pd.read_csv('./data/registrar_domains_2014-2016.csv')
dfRegistrarAllDomains.registrarname.fillna('unknown', inplace=True)

dfRegistrarLegacyDomains = pd.read_csv('./data/registrar_legacy_domains_2014-2016.csv')
dfRegistrarLegacyDomains.registrarname.fillna('unknown', inplace=True)

dfRegistrarStopBadwareDomains = pd.read_csv('./data/domains-per-registrar-stopbadware_2014-2016.csv')
dfRegistrarStopBadwareDomains.registrarname.fillna('unknown', inplace=True)

dfRegistrarStopBadwareLegacyDomains = pd.read_csv('./data/domains-legacy-per-registrar-stopbadware-2014-2016.csv')
dfRegistrarStopBadwareLegacyDomains.registrarname.fillna('unknown', inplace=True)


#2
#1: remove company types and icann registrar ids
#2: only keep chars in a-z
regex = re.compile('\(.*\)|LLC|INC|LTD|CORP|gmbh|limited|[^a-zA-Z]',  re.IGNORECASE)

def normalizeRegistrar(df):
   for i, row in df.iterrows():
      registrar = row['registrarname']
      if not registrar:
         registrar = "unknown"
      else:
         registrar = registrar.lower()
        
      registrar = regex.sub('', registrar)
      df.ix[i, 'registrarname'] = registrar

normalizeRegistrar(dfRegistrarAllDomains)
normalizeRegistrar(dfRegistrarLegacyDomains)
normalizeRegistrar(dfRegistrarStopBadwareDomains)
normalizeRegistrar(dfRegistrarStopBadwareLegacyDomains)

#3
dfRegistrarAllDomains = dfRegistrarAllDomains.set_index(['registrarname'])
dfRegistrarLegacyDomains = dfRegistrarLegacyDomains.set_index(['registrarname'])
dfRegistrarStopBadwareDomains = dfRegistrarStopBadwareDomains.set_index(['registrarname'])
dfRegistrarStopBadwareLegacyDomains = dfRegistrarStopBadwareLegacyDomains.set_index(['registrarname'])

#4
dfRegistrarAllDomains = dfRegistrarAllDomains.groupby(dfRegistrarAllDomains.index).sum()
dfRegistrarAllDomains = dfRegistrarAllDomains.sort_values(by="domains", ascending=False)

dfRegistrarLegacyDomains = dfRegistrarLegacyDomains.groupby(dfRegistrarLegacyDomains.index).sum()
dfRegistrarLegacyDomains = dfRegistrarLegacyDomains.sort_values(by="domains", ascending=False)

dfRegistrarStopBadwareDomains = dfRegistrarStopBadwareDomains.groupby(dfRegistrarStopBadwareDomains.index).sum()
dfRegistrarStopBadwareDomains = dfRegistrarStopBadwareDomains.sort_values(by="domains", ascending=False)

dfRegistrarStopBadwareLegacyDomains = dfRegistrarStopBadwareLegacyDomains.groupby(dfRegistrarStopBadwareLegacyDomains.index).sum()
dfRegistrarStopBadwareLegacyDomains = dfRegistrarStopBadwareLegacyDomains.sort_values(by="domains", ascending=False)
    
```

```python
dfRegistrarStopBadwareDomains[:10]
```

```python
dfCombined = dfRegistrarAllDomains
dfCombined['total_legacy'] = dfRegistrarLegacyDomains['domains']
dfCombined['total_newgtld'] = dfCombined['domains'] - dfCombined['total_legacy']
dfCombined['total_sb'] = dfRegistrarStopBadwareDomains['domains']
dfCombined['legacy_sb'] = dfRegistrarStopBadwareLegacyDomains['domains']
dfCombined['newgtld_sb'] = dfCombined['total_sb'] - dfCombined['legacy_sb']
dfCombined['sb_ratio'] = (dfCombined['total_sb'] / dfCombined['domains'])*100
dfCombined['sb_newgtld_ratio'] = (dfCombined['newgtld_sb'] / dfCombined['total_sb'])*100
dfCombined['sb_newgtld_2_newgtld'] = (dfCombined['newgtld_sb'] / dfCombined['total_newgtld'])*100
dfCombined['sb_legacy_2_legacy'] = (dfCombined['legacy_sb'] / dfCombined['total_legacy'])*100
```

```python
dfCombined.sort_values(by='domains', inplace=True, ascending=False)
```

```python
dfCombined[dfCombined["domains"] > 1000] [:10]
```

```python
# apwg domains / # all domains
dfCombined.sort_values(by='sb_ratio', inplace=True, ascending=False)
dfCombined[dfCombined["domains"] > 1000][["domains","total_sb", "sb_ratio"]] [:10]
```

```python
#apwg new gTLD domains / # all new gTLD domains 
dfCombined.sort_values(by='sb_newgtld_2_newgtld', inplace=True, ascending=False)
dfCombined[dfCombined["total_newgtld"] > 1000][["total_newgtld", "newgtld_sb", "sb_newgtld_2_newgtld"]] [:10]
```

```python
# apwg old gTLD domains / # all old gTLD domains 
dfCombined.sort_values(by='sb_legacy_2_legacy', inplace=True, ascending=False)
dfCombined[dfCombined["total_legacy"] > 1000][["total_legacy","legacy_sb", "sb_legacy_2_legacy"]] [:10]  
```

```python
font = {'family' : 'Helvetica',
        'size'   : 16}

matplotlib.rc('font', **font)

SMALL_SIZE = 16
plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=SMALL_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=SMALL_SIZE)    # legend fontsize
plt.rc('figure', titlesize=SMALL_SIZE)  # fontsize of the figure title

plt.figure(1)
fig = plt.figure(figsize=(15, 6), dpi=300)
ax = plt.subplot(121)
plt.xlabel('Old domain incidents',fontsize=24)
plt.ylabel('Registrars', fontsize=24)

#create cdf for all registrars with stopbadware legacy incidents
dfCombined.sort_values(by='legacy_sb', inplace=True, ascending=True)
dfValues = dfCombined[dfCombined["legacy_sb"] > 0][["legacy_sb"]]
sorted = np.sort(dfValues.values)
yvals = np.arange(len(sorted))/float(len(sorted))
plt.xscale('log')
plt.plot(sorted, yvals, linewidth=4.0)

ax = plt.subplot(122)
plt.xlabel('New gTLD domain incidents', fontsize=24)
plt.ylabel('Registrars',fontsize=24)


#create cdf for all registrars with stopbadware legacy incidents
dfCombined.sort_values(by='newgtld_sb', inplace=True, ascending=True)
dfValues = dfCombined[dfCombined["newgtld_sb"] > 0][["newgtld_sb"]]
sorted = np.sort(dfValues.values)
yvals = np.arange(len(sorted))/float(len(sorted))
plt.xscale('log')
plt.plot(sorted, yvals, linewidth=4.0)

plt.show()

fig.savefig('./SBW-incidents-ECDF.pdf')
```
