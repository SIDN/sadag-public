```python
import pandas as pd
import numpy as np
from __future__ import division
    
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import re
pd.set_option('display.mpl_style', 'default')
```

```python
def processData(inFileNew, inFileLegacy, outFile):
    dfNewTotal = pd.read_csv('./registrar-newgtld-total-count.csv', header=None, names=["ianaid","registrar","domains"], index_col="ianaid")
    dfLegacyTotal = pd.read_csv('./registrar-legacygtld-total-count.csv', header=None, names=["ianaid","registrar","domains"], index_col="ianaid")

    dfNewgTLD = pd.read_csv(inFileNew, index_col="ianaid", names=["ianaid","registrar","domains"])
    dfNewTotal["abused_new_gtld_count"] = dfNewgTLD["domains"]
    dfNewTotal["new_gtld_rate"] = (dfNewTotal["abused_new_gtld_count"] / dfNewTotal["domains"]) * 100.0
    dfNewTotal = dfNewTotal.sort_values(by="new_gtld_rate", ascending=False)
    dfNewTotal = dfNewTotal[dfNewTotal["abused_new_gtld_count"] > 1]
    dfNewTotal = dfNewTotal[dfNewTotal["domains"] > 100] 
    #dfNewTotal = dfNewTotal[["registrar","domains","abused_new_gtld_count","new_gtld_rate"]]
    dfNewTotal['domains'] = dfNewTotal['domains'].apply(lambda x: '{0:,d}'.format(x))
    dfNewTotal["abused_new_gtld_count"] = dfNewTotal["abused_new_gtld_count"].astype(int)
    dfNewTotal['abused_new_gtld_count'] = dfNewTotal['abused_new_gtld_count'].apply(lambda x: '{0:,d}'.format(x))
    dfNewTotal.reset_index(inplace=True)

    dfLegacygTLD = pd.read_csv(inFileLegacy, index_col="ianaid", names=["ianaid","registrar","domains"])
    dfLegacyTotal["abused_legacy_gtld_count"] = dfLegacygTLD["domains"]
    dfLegacyTotal["legacy_gtld_rate"] = (dfLegacyTotal["abused_legacy_gtld_count"] / dfLegacyTotal["domains"]) * 100.0
    dfLegacyTotal = dfLegacyTotal.sort_values(by="legacy_gtld_rate", ascending=False)
    dfLegacyTotal = dfLegacyTotal[dfLegacyTotal["abused_legacy_gtld_count"] > 1]
    dfLegacyTotal = dfLegacyTotal[dfLegacyTotal["domains"] > 100] 
    #dfLegacyTotal = dfLegacyTotal[["registrar","domains","abused_legacy_gtld_count","legacy_gtld_rate"]]
    dfLegacyTotal['domains'] = dfLegacyTotal['domains'].apply(lambda x: '{0:,d}'.format(x))
    dfLegacyTotal["abused_legacy_gtld_count"] = dfLegacyTotal["abused_legacy_gtld_count"].astype(int)
    dfLegacyTotal['abused_legacy_gtld_count'] = dfLegacyTotal['abused_legacy_gtld_count'].apply(lambda x: '{0:,d}'.format(x))
    dfLegacyTotal.reset_index(inplace=True)

    # dfResult[["registrar","domains","abused_new_gtld_count","new_gtld_rate","abused_legacy_gtld_count","legacy_gtld_rate"]]
    dfResult = pd.concat([dfNewTotal, dfLegacyTotal], axis=1)
    #dfResult = dfResult[(dfResult["abused_new_gtld_count"] > 1)][0:10]
    dfResult.index += 1 
    dfResult = dfResult[0:10]
    dfResult.fillna(value=0, inplace=True)
    dfResult["ianaid"] = dfResult["ianaid"].astype(int)
    #dfResult["domains"] = dfResult["domains"].astype(int)
    #dfResult["abused_new_gtld_count"] = dfResult["abused_new_gtld_count"].astype(int)
    #dfResult["abused_legacy_gtld_count"] = dfResult["abused_legacy_gtld_count"].astype(int)
    dfResult = dfResult.round(2)
    dfResult.drop("ianaid", inplace=True, axis=1)
    dfResult.to_csv(outFile, sep='&')
    return dfResult
```
